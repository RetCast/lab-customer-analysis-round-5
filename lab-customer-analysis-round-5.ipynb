```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
df.head()
```

```python
numerical_df = df[df.select_dtypes('number').columns]
numerical_df
```

### - Splitting into X and Y

```python
y = numerical_df['Total Claim Amount']
y
```

```python
X = numerical_df.drop(['Total Claim Amount'], axis=1)
X
```

#### - Train Test split

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

#### - Normalization

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt = PowerTransformer(method='yeo-johnson', standardize=True)
pt.fit(X_train)
X_train_pt = pt.transform(X_train)
df_X_train_pt = pd.DataFrame(X_train_pt, columns=X_train.columns)
df_X_train_pt.head(3)
```

```python
X_test_pt = pt.transform(X_test)
df_X_test_pt = pd.DataFrame(X_test_pt, columns=X_train.columns)
df_X_test_pt.head(3)
```
